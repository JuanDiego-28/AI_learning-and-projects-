```python
# MAP function 

import numpy as np 

data3= pd.read_csv("/home/jdiego/Documentos/AI_projects/employee.csv")


data3.head()
```

```python
# function to diveide by ten 
def by_ten (x): 
    return x/10 
```

```python
data3["DailyRate"] = data3["DailyRate"].map(by_ten)

data3["DailyRate"]   # map only works with series, not DataFrames 
```

```python
# APPLY works fro dataframes

data3[["DailyRate" , "MonthlyRate" ]].apply(by_ten) 

data3[["DailyRate" , "MonthlyRate" ]]
```

```python
# is netter to use lamnbda functions 

data3[["DailyRate" , "MonthlyRate" ]].apply(lambda x: x/10 ) 

data3[["DailyRate" , "MonthlyRate" ]]
```
